## Merge multiple variables into the same dataset

```python
vars_to_merge = [
    "e3t",
    "thetao_con",
    "so_abs",
    "zos",
    "zossq",
    "friver",
    "hfds",
    "rsdo",
    "mlotst",
    "mlotstsq",
    "mlotstmax",
    "mlotstmin",
    "ficeberg",
    "berg_latent_heat_flux",
    "pbo",
    "tos_con",
]
ds = xr.merge([catalog.eorca025_era5(var=var).to_dask() for var in vars_to_merge])
ds
```
